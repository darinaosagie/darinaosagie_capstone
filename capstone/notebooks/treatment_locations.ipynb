```python
import pandas as pd
import regex as re
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
treatment_locations= pd.read_csv("C:/Users/darin/Documents/NSS/darina_capstone/capstone/data/treatment_centers_location.csv")
treatment_locations
```

```python
m = folium.Map(location=[36.163248089975404, -86.77152505919005], zoom_start=8)
```

```python
for index, row in treatment_locations.iterrows():
    lat = row['LAT']
    lng = row['LON']
    center = row['Treament Center Name']
    popup = f"{center}<br>{row['Street Address']}<br>{row['City']}, {row['State']} {row['Zipcode']}"
    folium.Marker(location=[lat, lng], popup=popup, icon=folium.Icon(color='red')).add_to(m)

m.save("treatment_centers_map.html")
```
